```python
import osmnx as ox

# Define the coordinate as (latitude, longitude)
center_point = (10.991343783982689, 77.0044269727586)

# Download road network around the coordinate (radius in meters)
G_directed = ox.graph_from_point(center_point, dist=100, network_type="drive")

# FIX: Use the new convert module for undirected conversion
gu = ox.convert.to_undirected(G_directed)

```

```python
for i in gu.nodes:
    print(gu.nodes[i]['x'], gu.nodes[i]['y'])
```

```python
# convert coordinates to geojson compatible format
base = {
  "type": "FeatureCollection",
  "features": []
}

for i in gu.nodes:
    # Create a new feature dict for each node (don't reuse one dict!)
    base["features"].append({
        "type": "Feature",
        "properties": {
            "id": i
        },
        "geometry": {
            "type": "Point",
            "coordinates": [gu.nodes[i]["x"], gu.nodes[i]["y"]]
        }
    })


```

```python
for u,v,k,d in gu.edges(data=True, keys=True):
    print(u,v,k,d)
    print(d.keys())
```
